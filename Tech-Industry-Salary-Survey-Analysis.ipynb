```python
import pandas as pd #Load and inspect data
df = pd.read_csv("Tech_Professionals_Survey(1).csv")
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['Browser', 'OS', 'City', 'Country', 'Referrer', 'Email'], inplace=True) #Drop irrevelant Column
```

```python
df
```

```python
df = df.dropna(subset=['Salary_USD'])  # Drop rows with no salary
df.fillna('Unknown', inplace=True)
```

```python
df['Salary_USD'] = pd.to_numeric(df['Salary_USD'], errors='coerce')
```

```python
print(df.columns)
```

```python
print("Basic Salary Stats:")
print(df['Salary_USD'].describe())
```

```python
df
```

```python
def get_mid_salary(s):
    try:
        if pd.isnull(s):
            return None
        s = s.lower().replace('k', '').strip()  # remove 'k' and extra spaces
        if '-' in s:
            low, high = s.split('-')
            return (int(low) + int(high)) * 500  # midpoint * 1000 / 2
        else:
            return None  # not a range, so skip
    except:
        return None
    
df['salary'] = df['Salary_USD'].apply(get_mid_salary)
```

```python
df
```

```python
avg_salary_by_role = df.groupby('Role')['salary'].mean()

df
```

```python
avg_salary_by_role = df.groupby('Role')['salary'].mean()
top_roles = avg_salary_by_role.sort_values(ascending=False).head(10)
print(top_roles)
```
